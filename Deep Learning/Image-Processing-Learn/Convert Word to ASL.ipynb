```python
import os.path
from pathlib import Path 
import pandas as pd 
from PIL import Image
import glob
```

```python
def get_paths_labels(path, allowed_extension="jpg"):
        global Path
        images_dir = Path(path)
        print(images_dir)
        filepaths = pd.Series((images_dir.glob(fr'**/*.{allowed_extension}'))).astype(str)
        filepaths.name = "path"
        
        labels = filepaths.str.split("\\")[:].str[-2]
        labels.name = "label"

        # Concatenate filepaths and labels
        df = pd.concat([filepaths, labels], axis=1)

        # Shuffle the DataFrame and reset index
        df = df.sample(frac=1).reset_index(drop = True)
        return df
```

```python
def wordValitaion(wordCheck):
    paths = []
    words = list(wordCheck.upper().replace(" ",""))
    for word in words:
        paths.append(df[df['path'].str.contains(word+"_test")]['path'].item())
        
    return paths
```

```python
def imageList(paths):
    image_list = []
    for path in paths:
        for filename in glob.glob(path):
            im=Image.open(filename)
            image_list.append(im)
    return image_list
    
```
